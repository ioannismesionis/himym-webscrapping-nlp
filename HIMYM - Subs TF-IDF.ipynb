```python
import zipfile
```

```python
# FUNCTION TO SAVE THE CONTENT OF EVERY SUBTITLE
def subReader(path_file):
    subtitles = []  # LIST TO SAVE THE SUBTITLES
    zipFile = zipfile.ZipFile(path_file)
    for file in zipFile.infolist():
        sub = zipFile.open(file).readlines()
        subtitles.append(sub)
        
    return(subtitles)
```
