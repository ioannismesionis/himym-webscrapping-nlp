```python
## LOAD PACKAGES
from selenium import webdriver 
from selenium.webdriver.common.by import By 
from selenium.webdriver.support.ui import WebDriverWait 
from selenium.webdriver.support import expected_conditions as EC 
from selenium.common.exceptions import TimeoutException

from bs4 import BeautifulSoup
import pandas as pd
```

```python
## OPEN THE CHROME BROWSER CONTROLLED BY WEBDRIVER
browser = webdriver.Chrome('C:/Users/ioannisM/Downloads/chromedriver')
```

```python
## GET THE WEBSITE
browser.get("https://www.imdb.com/chart/top")
```

```python
my_movies = browser.find_element_by_class_name('lister-list')
```

```python
my_movies_data = BeautifulSoup(my_movies.get_attribute('outerHTML'), 'lxml')
my_movies_data
```

```python
titles_raw = my_movies_data.find_all('td', attrs = {'class':'titleColumn'})
```

```python
titles = [titles_raw[i].a.text for i in range(0, len(titles_raw))]
titles
```

```python
ratings_raw = my_movies_data.find_all('td', attrs = {'class':'imdbRating'})
```

```python
ratings = [float(ratings_raw[i].text.replace("\n", '')) for i in range(0, len(titles_raw))]
ratings
```

```python
data = pd.DataFrame(zip(titles, ratings))
```

```python
data.columns = ["Titles", "Ratings"]
```
